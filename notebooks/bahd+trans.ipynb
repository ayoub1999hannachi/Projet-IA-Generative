```python
import json
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from tqdm.auto import tqdm

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300

# ---------------------------
# SETTINGS
# ---------------------------
INTENTS_PATH = "/content/intents.json"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("="*70)
print(f"{'CONFIGURATION':^70}")
print("="*70)
print(f"Device: {DEVICE}")
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA Device: {torch.cuda.get_device_name(0)}")
print("="*70)

# ---------------------------
# 0) Load intents.json -> DataFrame
# ---------------------------
print("\n" + "="*70)
print(f"{'DATA LOADING':^70}")
print("="*70)

with open(INTENTS_PATH, "r", encoding="utf-8") as f:
    intents = json.load(f)

pairs = []
for intent in intents.get("intents", intents.get("intents", [])) if isinstance(intents, dict) else intents:
    tag = intent.get("tag")
    patterns = intent.get("patterns", [])
    responses = intent.get("responses", [])
    if len(responses) == 0:
        continue
    resp = responses[0]
    for p in patterns:
        pairs.append({"question": p, "answer": resp, "tag": tag})

df = pd.DataFrame(pairs)
print(f"Total pairs extracted: {len(df)}")
print(f"Number of unique tags: {df['tag'].nunique()}")
print(f"\nDataset Statistics:")
print(f"  - Mean question length: {df['question'].str.split().str.len().mean():.2f} words")
print(f"  - Mean answer length: {df['answer'].str.split().str.len().mean():.2f} words")
print(f"  - Max question length: {df['question'].str.split().str.len().max()} words")
print(f"  - Max answer length: {df['answer'].str.split().str.len().max()} words")

print("\nTop 10 Most Common Tags:")
print(df['tag'].value_counts().head(10))

print("\nSample Data:")
display(df.head(8))

# ---------------------------
# 1) Tokenizer + Vocab Builder
# ---------------------------
print("\n" + "="*70)
print(f"{'VOCABULARY BUILDING':^70}")
print("="*70)

def basic_tokenizer(text):
    if isinstance(text, str):
        return text.lower().strip().split()
    return []

def build_vocab(texts, min_freq=1):
    freq = {}
    for sent in texts:
        for tok in basic_tokenizer(sent):
            freq[tok] = freq.get(tok, 0) + 1

    itos = ["<pad>", "<unk>", "<start>", "<end>"]
    for tok, f in freq.items():
        if f >= min_freq:
            itos.append(tok)

    stoi = {tok: i for i, tok in enumerate(itos)}
    return stoi, itos

PAD = "<pad>"
UNK = "<unk>"
START = "<start>"
END = "<end>"

all_text = list(df["question"]) + list(df["answer"])
stoi, itos = build_vocab(all_text, min_freq=1)

PAD_IDX = stoi[PAD]
UNK_IDX = stoi[UNK]
START_IDX = stoi[START]
END_IDX = stoi[END]

vocab_size = len(itos)
print(f"Vocabulary size: {vocab_size}")
print(f"Special tokens:")
print(f"  - PAD: {PAD_IDX}")
print(f"  - UNK: {UNK_IDX}")
print(f"  - START: {START_IDX}")
print(f"  - END: {END_IDX}")

# Vocabulary statistics
word_freq = {}
for sent in all_text:
    for tok in basic_tokenizer(sent):
        word_freq[tok] = word_freq.get(tok, 0) + 1

print(f"\nVocabulary Statistics:")
print(f"  - Unique words: {len(word_freq)}")
print(f"  - Most common words: {sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:10]}")

# ---------------------------
# 2) Numericalize Helper
# ---------------------------
def numericalize(text):
    tokens = [START] + basic_tokenizer(text) + [END]
    ids = [stoi.get(tok, UNK_IDX) for tok in tokens]
    return torch.tensor(ids, dtype=torch.long)

# ---------------------------
# 3) Dataset + Collate
# ---------------------------
class QADataset(Dataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        q = self.df.iloc[idx]["question"]
        a = self.df.iloc[idx]["answer"]
        return numericalize(q), numericalize(a)

def collate_fn(batch):
    src, tgt = zip(*batch)
    src_lens = torch.tensor([len(s) for s in src])
    tgt_lens = torch.tensor([len(t) for t in tgt])
    src_pad = pad_sequence(src, batch_first=True, padding_value=PAD_IDX)
    tgt_pad = pad_sequence(tgt, batch_first=True, padding_value=PAD_IDX)
    return src_pad, src_lens, tgt_pad, tgt_lens

train_df, val_df = train_test_split(df, test_size=0.15, random_state=42)
train_loader = DataLoader(QADataset(train_df), batch_size=64, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(QADataset(val_df), batch_size=64, shuffle=False, collate_fn=collate_fn)

print("\n" + "="*70)
print(f"{'DATA SPLIT':^70}")
print("="*70)
print(f"Training samples: {len(train_df)}")
print(f"Validation samples: {len(val_df)}")
print(f"Batch size: 64")
print(f"Number of training batches: {len(train_loader)}")
print(f"Number of validation batches: {len(val_loader)}")

# ---------------------------
# 4) Model Architecture
# ---------------------------
print("\n" + "="*70)
print(f"{'MODEL ARCHITECTURE':^70}")
print("="*70)

class Encoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hid_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=PAD_IDX)
        self.rnn = nn.LSTM(emb_dim, hid_dim, batch_first=True, bidirectional=True)
        self.fc_h = nn.Linear(hid_dim * 2, hid_dim)
        self.fc_c = nn.Linear(hid_dim * 2, hid_dim)

    def forward(self, src, src_len):
        emb = self.embedding(src)
        packed = pack_padded_sequence(emb, src_len.cpu(), batch_first=True, enforce_sorted=False)
        output, (h, c) = self.rnn(packed)
        output, _ = pad_packed_sequence(output, batch_first=True)

        h_cat = torch.cat((h[-2], h[-1]), dim=1)
        c_cat = torch.cat((c[-2], c[-1]), dim=1)

        h_dec = torch.tanh(self.fc_h(h_cat)).unsqueeze(0)
        c_dec = torch.tanh(self.fc_c(c_cat)).unsqueeze(0)

        return output, (h_dec, c_dec)

class BahdanauAttention(nn.Module):
    def __init__(self, enc_hid, dec_hid):
        super().__init__()
        self.attn = nn.Linear(enc_hid*2 + dec_hid, dec_hid)
        self.v = nn.Linear(dec_hid, 1, bias=False)

    def forward(self, hidden, enc_outputs, mask):
        h = hidden[0].permute(1,0,2)
        h = h.repeat(1, enc_outputs.size(1), 1)
        e = torch.tanh(self.attn(torch.cat((h, enc_outputs), dim=2)))
        scores = self.v(e).squeeze(2)
        scores = scores.masked_fill(mask==0, -1e9)
        return torch.softmax(scores, dim=1)

class Decoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, enc_hid, dec_hid):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=PAD_IDX)
        self.attn = BahdanauAttention(enc_hid, dec_hid)
        self.rnn = nn.LSTM(enc_hid*2 + emb_dim, dec_hid, batch_first=True)
        self.fc = nn.Linear(enc_hid*2 + dec_hid + emb_dim, vocab_size)

    def forward(self, input_tok, hidden, enc_outputs, mask):
        emb = self.embedding(input_tok.unsqueeze(1))
        attn = self.attn(hidden, enc_outputs, mask).unsqueeze(1)
        context = attn @ enc_outputs
        rnn_input = torch.cat((emb, context), dim=2)
        output, hidden = self.rnn(rnn_input, hidden)
        output = output.squeeze(1)
        context = context.squeeze(1)
        emb = emb.squeeze(1)
        pred = self.fc(torch.cat((output, context, emb), dim=1))
        return pred, hidden, attn.squeeze(1)

EMB_DIM = 256
HID_DIM = 256
enc = Encoder(vocab_size, EMB_DIM, HID_DIM).to(DEVICE)
dec = Decoder(vocab_size, EMB_DIM, HID_DIM, HID_DIM).to(DEVICE)

# Model summary
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"\nEncoder Architecture:")
print(f"  - Embedding dimension: {EMB_DIM}")
print(f"  - Hidden dimension: {HID_DIM}")
print(f"  - Bidirectional LSTM: Yes")
print(f"  - Trainable parameters: {count_parameters(enc):,}")

print(f"\nDecoder Architecture:")
print(f"  - Attention mechanism: Bahdanau (Additive)")
print(f"  - Hidden dimension: {HID_DIM}")
print(f"  - Trainable parameters: {count_parameters(dec):,}")

print(f"\nTotal trainable parameters: {count_parameters(enc) + count_parameters(dec):,}")

criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)
optimizer = torch.optim.Adam(list(enc.parameters()) + list(dec.parameters()), lr=1e-3)

# ---------------------------
# 5) Training with Enhanced Metrics
# ---------------------------
train_losses, val_losses = [], []
train_accs, val_accs = [], []
train_perplexities, val_perplexities = [], []
learning_rates = []

def create_mask(src):
    return (src != PAD_IDX)

def accuracy(preds, target, ignore_index=PAD_IDX):
    preds = preds.argmax(dim=-1)
    mask = target != ignore_index
    correct = ((preds == target) & mask).sum().item()
    total = mask.sum().item()
    return correct / total if total > 0 else 0

def train_epoch():
    enc.train(); dec.train()
    total_loss, total_acc, num_tokens = 0.0, 0.0, 0

    progress_bar = tqdm(train_loader, desc="Training", leave=False)
    for src, src_lens, tgt, tgt_lens in progress_bar:
        src, src_lens, tgt = src.to(DEVICE), src_lens.to(DEVICE), tgt.to(DEVICE)
        optimizer.zero_grad()
        enc_outputs, hidden = enc(src, src_lens)
        mask = create_mask(src)

        input_tok = tgt[:,0]
        preds_all = []

        for t in range(1, tgt.size(1)):
            preds, hidden, _ = dec(input_tok, hidden, enc_outputs, mask)
            preds_all.append(preds)
            if torch.rand(1).item() < 0.5:
                input_tok = tgt[:,t]
            else:
                input_tok = preds.argmax(1)

        preds_all = torch.stack(preds_all, dim=1)
        loss = criterion(preds_all.reshape(-1, vocab_size), tgt[:,1:].reshape(-1))
        acc = accuracy(preds_all, tgt[:,1:])
        loss.backward()
        torch.nn.utils.clip_grad_norm_(list(enc.parameters()) + list(dec.parameters()), 1.0)
        optimizer.step()

        total_loss += loss.item()
        total_acc += acc
        num_tokens += (tgt[:,1:] != PAD_IDX).sum().item()

        progress_bar.set_postfix({'loss': loss.item(), 'acc': acc})

    avg_loss = total_loss / len(train_loader)
    avg_acc = total_acc / len(train_loader)
    perplexity = np.exp(avg_loss)

    return avg_loss, avg_acc, perplexity

def eval_epoch():
    enc.eval(); dec.eval()
    total_loss, total_acc, num_tokens = 0.0, 0.0, 0

    progress_bar = tqdm(val_loader, desc="Validation", leave=False)
    with torch.no_grad():
        for src, src_lens, tgt, tgt_lens in progress_bar:
            src, src_lens, tgt = src.to(DEVICE), src_lens.to(DEVICE), tgt.to(DEVICE)
            enc_outputs, hidden = enc(src, src_lens)
            mask = create_mask(src)
            input_tok = tgt[:,0]
            preds_all = []
            for t in range(1, tgt.size(1)):
                preds, hidden, _ = dec(input_tok, hidden, enc_outputs, mask)
                preds_all.append(preds)
                input_tok = tgt[:,t]
            preds_all = torch.stack(preds_all, dim=1)
            loss = criterion(preds_all.reshape(-1, vocab_size), tgt[:,1:].reshape(-1))
            acc = accuracy(preds_all, tgt[:,1:])
            total_loss += loss.item()
            total_acc += acc
            num_tokens += (tgt[:,1:] != PAD_IDX).sum().item()

            progress_bar.set_postfix({'loss': loss.item(), 'acc': acc})

    avg_loss = total_loss / len(val_loader)
    avg_acc = total_acc / len(val_loader)
    perplexity = np.exp(avg_loss)

    return avg_loss, avg_acc, perplexity

print("\n" + "="*70)
print(f"{'TRAINING':^70}")
print("="*70)

EPOCHS = 30
best_val_loss = float('inf')
patience = 5
patience_counter = 0

for epoch in range(EPOCHS):
    train_loss, train_acc, train_ppl = train_epoch()
    val_loss, val_acc, val_ppl = eval_epoch()

    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_perplexities.append(train_ppl)
    val_perplexities.append(val_ppl)
    learning_rates.append(optimizer.param_groups[0]['lr'])

    print(f"Epoch {epoch+1:02d}/{EPOCHS}  "
          f"Train Loss: {train_loss:.4f} Acc: {train_acc:.3f} PPL: {train_ppl:.2f}  |  "
          f"Val Loss: {val_loss:.4f} Acc: {val_acc:.3f} PPL: {val_ppl:.2f}")

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'encoder_state_dict': enc.state_dict(),
            'decoder_state_dict': dec.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'train_loss': train_loss,
            'val_loss': val_loss,
        }, 'best_model.pt')
    else:
        patience_counter += 1

    if patience_counter >= patience:
        print(f"\nEarly stopping triggered at epoch {epoch+1}")
        break

# Save detailed training history
hist = pd.DataFrame({
    "epoch": range(1, len(train_losses) + 1),
    "train_loss": train_losses,
    "val_loss": val_losses,
    "train_acc": train_accs,
    "val_acc": val_accs,
    "train_perplexity": train_perplexities,
    "val_perplexity": val_perplexities,
    "learning_rate": learning_rates
})
hist.to_csv("training_history_detailed.csv", index=False)
print("\n✓ Saved training_history_detailed.csv")

# ---------------------------
# 6) Comprehensive Visualization
# ---------------------------
print("\n" + "="*70)
print(f"{'VISUALIZATION':^70}")
print("="*70)

# Create comprehensive plots
fig = plt.figure(figsize=(20, 12))

# 1. Loss curves with moving average
ax1 = plt.subplot(3, 3, 1)
plt.plot(train_losses, label="Train Loss", alpha=0.6)
plt.plot(val_losses, label="Val Loss", alpha=0.6)
# Moving average
window = 5
if len(train_losses) >= window:
    train_ma = pd.Series(train_losses).rolling(window=window).mean()
    val_ma = pd.Series(val_losses).rolling(window=window).mean()
    plt.plot(train_ma, label=f"Train MA({window})", linewidth=2)
    plt.plot(val_ma, label=f"Val MA({window})", linewidth=2)
plt.title("Loss Curves with Moving Average", fontsize=12, fontweight='bold')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)

# 2. Accuracy curves
ax2 = plt.subplot(3, 3, 2)
plt.plot(train_accs, label="Train Acc", marker='o', markersize=3)
plt.plot(val_accs, label="Val Acc", marker='s', markersize=3)
plt.title("Accuracy Curves", fontsize=12, fontweight='bold')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True, alpha=0.3)

# 3. Perplexity curves
ax3 = plt.subplot(3, 3, 3)
plt.plot(train_perplexities, label="Train PPL")
plt.plot(val_perplexities, label="Val PPL")
plt.title("Perplexity Curves", fontsize=12, fontweight='bold')
plt.xlabel("Epoch")
plt.ylabel("Perplexity")
plt.legend()
plt.grid(True, alpha=0.3)
plt.yscale('log')

# 4. Train vs Val Loss scatter
ax4 = plt.subplot(3, 3, 4)
plt.scatter(train_losses, val_losses, alpha=0.6, s=50)
plt.plot([min(train_losses), max(train_losses)],
         [min(train_losses), max(train_losses)], 'r--', label='Perfect fit')
plt.title("Train vs Val Loss", fontsize=12, fontweight='bold')
plt.xlabel("Train Loss")
plt.ylabel("Val Loss")
plt.legend()
plt.grid(True, alpha=0.3)

# 5. Overfitting analysis
ax5 = plt.subplot(3, 3, 5)
gap = np.array(val_losses) - np.array(train_losses)
plt.plot(gap, color='red', linewidth=2)
plt.axhline(y=0, color='black', linestyle='--', alpha=0.3)
plt.title("Generalization Gap (Val - Train Loss)", fontsize=12, fontweight='bold')
plt.xlabel("Epoch")
plt.ylabel("Loss Difference")
plt.grid(True, alpha=0.3)

# 6. Learning rate schedule (if using scheduler)
ax6 = plt.subplot(3, 3, 6)
plt.plot(learning_rates, color='orange', linewidth=2)
plt.title("Learning Rate Schedule", fontsize=12, fontweight='bold')
plt.xlabel("Epoch")
plt.ylabel("Learning Rate")
plt.grid(True, alpha=0.3)

# 7. Class distribution
ax7 = plt.subplot(3, 3, 7)
tag_counts = train_df['tag'].value_counts().head(15)
tag_counts.plot(kind='barh', color='skyblue')
plt.title("Top 15 Intent Distributions (Train)", fontsize=12, fontweight='bold')
plt.xlabel("Count")
plt.ylabel("Tag")
plt.tight_layout()

# 8. Loss improvement per epoch
ax8 = plt.subplot(3, 3, 8)
train_improvement = -np.diff([train_losses[0]] + train_losses)
val_improvement = -np.diff([val_losses[0]] + val_losses)
plt.bar(range(len(train_improvement)), train_improvement, alpha=0.6, label='Train')
plt.bar(range(len(val_improvement)), val_improvement, alpha=0.6, label='Val')
plt.title("Loss Improvement per Epoch", fontsize=12, fontweight='bold')
plt.xlabel("Epoch")
plt.ylabel("Loss Decrease")
plt.legend()
plt.grid(True, alpha=0.3)

# 9. Metrics summary heatmap
ax9 = plt.subplot(3, 3, 9)
summary_data = np.array([
    train_losses[-5:],
    val_losses[-5:],
    train_accs[-5:],
    val_accs[-5:]
]).T
sns.heatmap(summary_data, annot=True, fmt='.3f',
            xticklabels=['Train Loss', 'Val Loss', 'Train Acc', 'Val Acc'],
            yticklabels=[f'E{len(train_losses)-4+i}' for i in range(5)],
            cmap='YlOrRd', ax=ax9)
plt.title("Last 5 Epochs Metrics", fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('training_analysis_comprehensive.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Saved training_analysis_comprehensive.png")

# ---------------------------
# 7) Inference and Evaluation
# ---------------------------
print("\n" + "="*70)
print(f"{'MODEL EVALUATION':^70}")
print("="*70)

def generate_answer(question, max_len=40, return_attn=False):
    enc.eval(); dec.eval()
    ids = numericalize(question).unsqueeze(0).to(DEVICE)
    length = torch.tensor([ids.size(1)], device=DEVICE)
    with torch.no_grad():
        enc_outputs, hidden = enc(ids, length)
        mask = create_mask(ids)
        input_tok = torch.tensor([START_IDX], device=DEVICE)
        output_tokens = []
        attn_weights = []
        for _ in range(max_len):
            preds, hidden, attn = dec(input_tok, hidden, enc_outputs, mask)
            top = preds.argmax(1).item()
            if top == END_IDX:
                break
            output_tokens.append(itos[top])
            attn_weights.append(attn.cpu().numpy()[0])
            input_tok = torch.tensor([top], device=DEVICE)
    if return_attn:
        return " ".join(output_tokens), np.stack(attn_weights, axis=0) if len(attn_weights)>0 else None
    return " ".join(output_tokens)

# Generate predictions on validation set
samples = val_df.sample(min(50, len(val_df)), random_state=1).reset_index(drop=True)
preds = []
for i, row in tqdm(samples.iterrows(), total=len(samples), desc="Generating predictions"):
    ans = generate_answer(row["question"])
    preds.append({
        "question": row["question"],
        "expected": row["answer"],
        "predicted": ans,
        "tag": row["tag"]
    })
pred_df = pd.DataFrame(preds)

print("\nSample Predictions:")
display(pred_df.head(20))

pred_df.to_csv("sample_predictions_detailed.csv", index=False)
print("\n✓ Saved sample_predictions_detailed.csv")

# ---------------------------
# 8) Summary Statistics Table
# ---------------------------
print("\n" + "="*70)
print(f"{'FINAL SUMMARY':^70}")
print("="*70)

summary_stats = {
    "Metric": [
        "Best Train Loss",
        "Best Val Loss",
        "Best Train Accuracy",
        "Best Val Accuracy",
        "Final Train Loss",
        "Final Val Loss",
        "Final Train Accuracy",
        "Final Val Accuracy",
        "Final Train Perplexity",
        "Final Val Perplexity",
        "Total Epochs",
        "Total Parameters",
        "Training Samples",
        "Validation Samples"
    ],
    "Value": [
        f"{min(train_losses):.4f}",
        f"{min(val_losses):.4f}",
        f"{max(train_accs):.4f}",
        f"{max(val_accs):.4f}",
        f"{train_losses[-1]:.4f}",
        f"{val_losses[-1]:.4f}",
        f"{train_accs[-1]:.4f}",
        f"{val_accs[-1]:.4f}",
        f"{train_perplexities[-1]:.2f}",
        f"{val_perplexities[-1]:.2f}",
        f"{len(train_losses)}",
        f"{count_parameters(enc) + count_parameters(dec):,}",
        f"{len(train_df):,}",
        f"{len(val_df):,}"
    ]
}

summary_df = pd.DataFrame(summary_stats)
print("\n" + summary_df.to_string(index=False))
summary_df.to_csv("training_summary.csv", index=False)
print("\n✓ Saved training_summary.csv")

# ---------------------------
# 9) Attention Visualization for Multiple Examples
# ---------------------------
print("\n" + "="*70)
print(f"{'ATTENTION VISUALIZATION':^70}")
print("="*70)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx in range(min(4, len(samples))):
    example_row = samples.iloc[idx]
    gen_ans, attn_matrix = generate_answer(example_row["question"], return_attn=True)

    ax = axes[idx]
    if attn_matrix is not None and len(attn_matrix) > 0:
        im = ax.imshow(attn_matrix.T, aspect='auto', origin='lower', cmap='viridis')
        ax.set_xlabel("Decoder Step", fontsize=10)
        ax.set_ylabel("Encoder Position", fontsize=10)
        ax.set_title(f"Example {idx+1}: {example_row['question'][:30]}...",
                    fontsize=10, fontweight='bold')

        # Add colorbar
        plt.colorbar(im, ax=ax, label="Attention Weight")

        # Set ticks
        enc_tokens = [START] + basic_tokenizer(example_row["question"]) + [END]
        if len(enc_tokens) <= 20:
            ax.set_yticks(range(len(enc_tokens)))
            ax.set_yticklabels(enc_tokens, fontsize=8)
    else:
        ax.text(0.5, 0.5, 'No attention data', ha='center', va='center')
        ax.axis('off')

plt.tight_layout()
plt.savefig('attention_heatmaps.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Saved attention_heatmaps.png")


```
